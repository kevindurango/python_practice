DISPLAYING MULTIPLE CHARTS

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('c:/Users/ACER/Documents/python_koica/data/iris.csv')

fig, axes = plt.subplots(nrows=2, ncols=2)

df['Petal_Length'].plot.hist(ax=axes[0, 0])
df['Petal_Length'].plot.box(ax=axes[0, 1])

fd = df['Species'].value_counts()
fd.plot.pie(ax=axes[1, 0])
fd.plot.bar(ax=axes[1, 1])

fig.suptitle('Multiple Graph Example', fontsize=16, fontweight='bold')

plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

hobbies = pd.Series(['Climbing', 'Fishing', 'Golf', 'Swimming',
                     'Climbing', 'Climbing', 'Fishing', 'Climbing',
                     'Fishing', 'Swimming','Golf'])

table = hobbies.value_counts()
print(table)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

fd = hobbies.value_counts()
fd.plot.pie(title='Hobby Distribution', ax=axes[0])
fd.plot.barh(title='Hobby Count', ax=axes[1])


fig.suptitle('Graph Example', fontsize=16, fontweight='bold')

plt.show()
```
