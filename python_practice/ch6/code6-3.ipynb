SCATTER PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('c:/Users/ACER/Documents/python_koica/data/iris.csv')
dict = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

colors = [dict[key] for key in df['Species']]
colors


df.plot.scatter(x='Petal_Length', y='Petal_Width', c=colors, s=20, marker='*')
plt.show()
```
