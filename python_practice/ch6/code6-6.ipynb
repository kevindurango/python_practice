```python
import pandas as pd
import matplotlib.pyplot as plt

late1 = [5,8,7,9,4,6,12,13,8,6,6,4]
late2 = [4,6,5,8,7,8,10,11,6,5,7,3]

dict = {'late1': late1, 'late2': late2}
late = pd.DataFrame(dict, index=list(range(1, 13)))

late.plot(title='late students per month', xlabel='Month', 
        ylabel='Number of Late Students', linestyle='dotted', marker='o')

plt.legend(loc='upper right')
plt.show()
```
